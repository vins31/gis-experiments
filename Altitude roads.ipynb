# Altitude roads

## Purpose
When you use Google Maps or opencyclemap to plan a trip in a plain, **everything looks flat**, even if you have 100m elevation changes on your way. If there is a small hill or a valley you may not see it well because the **map is so overloaded** that you cannot focus on the coutour lines.

The solution would be to have a map with **custom elevation information optimized for you are area**.
In a glimpse, you would be able to see:

* small elevation changes, small hills, small valleys
* the steepest roads to avoid
* the minimal climb longuest path from home

This Python notebook will enable you to generate an elevation map optimized for your area.

Beware, there are some parameters to tune.

## Code

```python
from credentials import osmapi_username, osm_api_password, mapquestapi_key
import itertools
import json
from matplotlib.collections import LineCollection
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from operator import itemgetter
import osmapi
import smopy
import urllib
%matplotlib inline
%load_ext pep8_magic
```

### Loading the map tile image

* Define the bounding box of your map with lat_min, lon_min, lat_max, lon_max
* Tune the **zoom** variable to your needs. Too low implies not enough details and a small image, too big implies too many tiles downladed (slow) and a big image.

```python
# Parameters
lat_min, lon_min = 43.622046, 1.604956
lat_max, lon_max = 43.703261, 1.711901
zoom = 14

# Load and display the map tile
img_map = smopy.Map((lat_min, lon_min, lat_max, lon_max), z=zoom)
img_map.show_ipython()
# Save the map for potential future use
img_map.save_png("map_%.2f_%.2f_%.2f_%f.png" %
                 (lat_min, lon_min, lat_max, lon_max))
```

### Retrieve and index OSM map elements

```python
api = osmapi.OsmApi(api="api.openstreetmap.org",
                    username=osmapi_username,
                    password=osm_api_password)

# Retrieve all Open Street Map data of the specified bounding box
# using the Open Street Map API
osm_map = api.Map(lon_min, lat_min, lon_max, lat_max)
# Save Open Street Map data for future use
with open("map_elements_%.2f_%.2f_%.2f_%f.json" %
          (lat_min, lon_min, lat_max, lon_max), "w+") as fd_osm_map:
    fd_osm_map.write(json.dumps(osm_map))
fd_osm_map.close()

# Extract the highways and index the nodes
# (the path of a way is defined by a list of nodes)
nodes = {}  # To retrieve a node by its id
highways = []
for element in osm_map:
    if element["type"] == "node":
        node = element["data"]
        nodes[node["id"]] = node
    elif element["type"] == "way":
        way = element["data"]
        if "highway" in way["tag"]:
            highways.append(way)
```

### Retrieve altitudes from MapQuest

```python
class RoadNodes:
    """Store the lists of the coordinates of the nodes of a given road"""
    def __init__(self, lx, ly, lz):
        self.lx = lx
        self.ly = ly
        self.lz = lz


def get_road_data(road):
    """For each node:
        1. compute its map coordinate from its (lat,lon)
        2. retrieve its altitude
        @arg road: OSM way element
        @return a RoadNodes instance"""
    lx, ly = [], []
    url_prefix = "http://open.mapquestapi.com/elevation/v1/profile?key=%s" \
        "&shapeFormat=raw&latLngCollection=" % mapquestapi_key
    url_suffix = ""
    for node_id in road["nd"]:
        node = nodes[node_id]
        coords = img_map.to_pixels(node["lat"], node["lon"])
        lx.append(coords[0])
        ly.append(coords[1])
        url_suffix += ",%s,%s" % (node["lat"], node["lon"])
    altitudes = json.loads(urllib.urlopen(url_prefix + url_suffix[1:]).read())
    lz = map(itemgetter("height"), altitudes["elevationProfile"])
    roadData = RoadNodes(lx, ly, lz)
    return roadData


# Retrieve the nodes information (x, y, altitude) of each road
# using Mapquest API and compute the minimal and maximal altitude of the map
roads_data = []
z_min = 9e99
z_max = -9e99
for road in highways:
    road_data = get_road_data(road)
    roads_data.append(road_data)
    z_min = min(z_min, min(road_data.lz))
    z_max = max(z_max, max(road_data.lz))
```

### Plot the map
You may want to change *figsize*, *colormap*, *linewidth* and *alpha*

```python
fig = plt.figure(figsize=(40, 25))
plt.hold(True)
plt.imshow(img_map.img, interpolation="bicubic")
colormap = plt.get_cmap('jet')
for road_data in roads_data:
    # Display the lines with lz values as color
    points = np.array([road_data.lx, road_data.ly]).T.reshape(-1, 1, 2)
    segments = np.concatenate([points[:-1], points[1:]], axis=1)
    lc = LineCollection(segments, array=np.asarray(road_data.lz),
                        cmap=colormap, norm=plt.Normalize(z_min, z_max),
                        linewidth=5, alpha=0.7)
    plt.gca().add_collection(lc)
plt.hold(False)
# Compute the colormap legend
ticks = np.linspace(z_min, z_max, 12)
cbar = fig.colorbar(lc, ticks=ticks)
cbar.ax.set_yticklabels(["%.2f m" % tick for tick in ticks])
plt.savefig("map_elevation_%.2f_%.2f_%.2f_%f.png"
            % (lat_min, lon_min, lat_max, lon_max))
```

In this elevation map you can noticed it exists a flat way to go from the Balerme lake to the Laragou lake (blue path).
